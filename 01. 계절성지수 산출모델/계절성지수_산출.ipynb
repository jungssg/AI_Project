# 2.
###[불량 데이터 처리]
### kopo_channel_seasonality_new.csv 자료를 담은   
### selloutData 변수에서
### QTY컬럼 음수(반품)인 경우 0, 양수인 경우 기존 QTY 값  
### 유지하는 로직을 적용하여 QTY_NEW 컬럼을 추가  


```python
url = '../dataset/kopo_channel_seasonality_new.csv'
selloutData = pd.read_csv(url)
selloutData
```

### np.where 사용 풀이

```python
import numpy as np

selloutData["QTY_NEW"] = np.where(selloutData.QTY < 0 , 0, selloutData.QTY)
selloutData.loc[ (selloutData.QTY < 0) & (selloutData.QTY_NEW != 0) ]
```

```python
selloutData
```

### loc 사용 풀이

```python
selloutData.loc[ selloutData.QTY < 0, "QTY_NEW" ] = 0
selloutData.loc[ selloutData.QTY >= 0, "QTY_NEW" ] = selloutData.QTY
selloutData
```

### 함수 사용 풀이

```python
def new_qty(qtyValue):
    """
    qytValue 를 입력받아서 0 보다 작으면 0, 0 이상이면 qtyValue값으로 qty_new값 할당
    """
    qtyNewValue = 0
    if qtyValue < 0:
        qtyNewValue = 0
    else:
       qtyNewValue = qtyValue
    return qtyNewValue
```

```python
selloutData["QTY_NEW"] = selloutData.QTY.apply(new_qty)
```

```python
selloutData
```

```python
### 검증
selloutData.loc[ (selloutData.QTY < 0) & (selloutData.QTY_NEW != 0) ]
```

### refinedSelloutData 부터 finalResult 까지 해보기

```python
refinedSelloutData
# 01. 데이터프레임 정렬하기 (REGIONID,PRODUCT,YEARWEEK) 기준
sortKey = ["REGIONID","PRODUCT","YEARWEEK"]
sortedData = refinedSelloutData.sort_values(by=sortKey, ignore_index=True)
sortedData

# 02. 그룹화할 데이터 ("REGIONID", "PRODUCT")
groupKey = ["REGIONID", "PRODUCT"]
groupData = sortedData.groupby(by=groupKey)
eachGroup = groupData.get_group( list(groupData.groups)[0])
eachGroup

# 03. 이동평균 구하기
eachGroup["MA5"] = eachGroup.QTY_NEW.rolling(window=5, center=True, min_periods=3).mean()
eachGroup

# 04. 함수화하기
def rollingMa5(eachGroup):
    # eachGroup = groupData.get_group( list(groupData.groups)[0])
    # 이동평균 구하기
    eachGroup["MA5_NEW"] = eachGroup.QTY_NEW.rolling(window=5, center=True, min_periods=3).mean()
    eachGroup["SEASONALITY"] = np.where(eachGroup.MA5_NEW == 0, 0, eachGroup.QTY_NEW / eachGroup.MA5_NEW)
    return eachGroup

# 05. 함수 적용하기
groupData = sortedData.groupby(by=groupkey).apply(rollingMa5)

# 06. 그룹인덱스 해제
groupData.reset_index(drop=True, inplace=True)
groupData
# 07. 그룹화하기 ("REGIONID", "PRODUCT", "WEEK")
groupKeys = ["REGIONID", "PRODUCT", "WEEK"]

# 08. finalResult
finalResult = groupData.groupby(by=groupKeys)["SEASONALITY"].agg(["mean"])
finalResult
# finalResult.to_csv('./season_result.csv')
```
